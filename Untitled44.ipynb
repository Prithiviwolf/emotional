````python
from transformers import pipeline

# Loading pretrained emotion detection model
print("Loading model... please wait ")

emotion_model = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=True
)

print("Model loaded successfully! ")
````

````python
def analyze_emotion(text):
    results = emotion_model(text)

    # Fix: results is already a list of dicts
    if isinstance(results[0], list):
        results = results[0]

    # Sort by score
    results = sorted(results, key=lambda x: x['score'], reverse=True)

    print(f"\n📝 Text: '{text}'")
    print(f"\n🎯 Top Emotion: {results[0]['label'].upper()} ({results[0]['score']*100:.1f}%)")
    print(f"\n📊 All Emotions:")

    for r in results:
        bar = "█" * int(r['score'] * 30)
        print(f"  {r['label']:<12} {bar} {r['score']*100:.1f}%")

# Test it
analyze_emotion("I just got my dream job offer today!")
print("\n" + "="*50)
analyze_emotion("I am so tired and nothing is going right for me.")
print("\n" + "="*50)
analyze_emotion("How dare they do this without telling me!")
````

````python
import matplotlib.pyplot as plt
import numpy as np

def analyze_and_plot(text):
    results = emotion_model(text)
    if isinstance(results[0], list):
        results = results[0]

    results = sorted(results, key=lambda x: x['score'], reverse=True)

    labels = [r['label'] for r in results]
    scores = [r['score'] * 100 for r in results]

    # Emotion colors
    colors = {
        'joy':      '#FFD700',
        'sadness':  '#4A90D9',
        'anger':    '#E74C3C',
        'fear':     '#8E44AD',
        'surprise': '#F39C12',
        'disgust':  '#27AE60',
        'neutral':  '#95A5A6'
    }

    bar_colors = [colors.get(l, '#3498DB') for l in labels]

    # Plot
    fig, ax = plt.subplots(figsize=(10, 5))
    bars = ax.barh(labels, scores, color=bar_colors, edgecolor='white', height=0.6)

    # Add value labels
    for bar, score in zip(bars, scores):
        ax.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2,
                f'{score:.1f}%', va='center', fontsize=12, fontweight='bold')

    ax.set_xlim(0, 115)
    ax.set_xlabel('Confidence Score (%)', fontsize=12)
    ax.set_title(f'🧠 BehaviorLens — Emotion Analysis\n"{text[:60]}..."'
                 if len(text) > 60 else f'🧠 BehaviorLens — Emotion Analysis\n"{text}"',
                 fontsize=13, fontweight='bold', pad=15)
    ax.invert_yaxis()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    plt.tight_layout()
    plt.show()

    print(f"\n🎯 Detected: {results[0]['label'].upper()} ({results[0]['score']*100:.1f}% confidence)")

# Test with your own sentence!
analyze_and_plot("I am nervous about my upcoming job interview")
````

````python
def analyze_personality(text):
    results = personality_model(text)

    if isinstance(results[0], list):
        results = results[0]

    # Big 5 trait full names
    trait_names = {
        'INFJ': 'Openness',
        'ENTP': 'Conscientiousness',
        'INTJ': 'Extraversion',
        'ENTJ': 'Agreeableness',
        'ISTP': 'Neuroticism'
    }

    results = sorted(results, key=lambda x: x['score'], reverse=True)

    print(f"\n📝 Text: '{text}'")
    print("\n🧠 Personality Traits:")

    for r in results:
        label = trait_names.get(r['label'], r['label'])
        bar = "█" * int(r['score'] * 30)
        print(f"  {label:<20} {bar} {r['score']*100:.1f}%")

    top = trait_names.get(results[0]['label'], results[0]['label'])
    score = results[0]['score'] * 100
    print(f"\n⭐ Dominant Trait: {top.upper()} ({score:.1f}%)")

# Test it!
analyze_personality("I love exploring new ideas and meeting new people every day!")
print("\n" + "="*50)
analyze_personality("I prefer working alone and planning everything carefully before I act.")
````

````python
def calculate_stress(emotion_dict):
    fear    = emotion_dict.get('fear', 0)
    sadness = emotion_dict.get('sadness', 0)
    anger   = emotion_dict.get('anger', 0)
    disgust = emotion_dict.get('disgust', 0)
    joy     = emotion_dict.get('joy', 0)
    neutral = emotion_dict.get('neutral', 0)

    # Stress formula
    stress = ((fear * 0.35) + (sadness * 0.25) +
              (anger * 0.25) + (disgust * 0.15) -
              (joy * 0.3) - (neutral * 0.1))

    # Clamp between 0 and 1
    stress = max(0, min(1, stress))
    stress_pct = round(stress * 100, 1)

    # Stress level label
    if stress_pct < 20:
        level = "🟢 Very Low"
    elif stress_pct < 40:
        level = "🟡 Low"
    elif stress_pct < 60:
        level = "🟠 Moderate"
    elif stress_pct < 80:
        level = "🔴 High"
    else:
        level = "🚨 Very High"

    return stress_pct, level

def generate_insight(top_emotion, top_trait, stress_pct, stress_level):
    insights = {
        ('joy', 'Extraversion'):      "This person is highly energetic and socially confident.",
        ('joy', 'Openness'):          "This person is creative, curious and in a positive state of mind.",
        ('joy', 'Agreeableness'):     "This person is warm, cooperative and easy to work with.",
        ('sadness', 'Neuroticism'):   "This person may be going through emotional difficulty. Handle with care.",
        ('anger', 'Neuroticism'):     "This person is frustrated and emotionally reactive right now.",
        ('fear', 'Neuroticism'):      "This person is anxious and may need reassurance.",
        ('neutral', 'Conscientiousness'): "This person is calm, focused and highly organized.",
        ('surprise', 'Openness'):     "This person is open to new experiences and easily excited.",
    }

    key = (top_emotion.lower(), top_trait)
    insight = insights.get(key,
              f"This person shows strong {top_trait} with {top_emotion} as dominant emotion.")

    if stress_pct > 60:
        insight += " ⚠️ High stress detected — may need support."
    elif stress_pct < 20:
        insight += " ✅ Overall mental state appears healthy and balanced."

    return insight

# Test it!
text = "I love exploring new ideas and meeting new people every day!"
results = emotion_model(text)
if isinstance(results[0], list):
    results = results[0]
results = sorted(results, key=lambda x: x['score'], reverse=True)
emotion_dict = {r['label']: r['score'] for r in results}

stress_pct, stress_level = calculate_stress(emotion_dict)
top_emotion = results[0]['label']

big5 = {
    'Openness':          round((emotion_dict.get('joy',0) * 0.4 + emotion_dict.get('surprise',0) * 0.4 + emotion_dict.get('neutral',0) * 0.2) * 100, 1),
    'Conscientiousness': round((emotion_dict.get('neutral',0) * 0.5 + emotion_dict.get('joy',0) * 0.3 + (1 - emotion_dict.get('sadness',0)) * 0.2) * 100, 1),
    'Extraversion':      round((emotion_dict.get('joy',0) * 0.6 + emotion_dict.get('surprise',0) * 0.2 + emotion_dict.get('neutral',0) * 0.2) * 100, 1),
    'Agreeableness':     round((emotion_dict.get('joy',0) * 0.4 + emotion_dict.get('neutral',0) * 0.3 + (1 - emotion_dict.get('sadness',0)) * 0.3) * 100, 1),
    'Neuroticism':       round((emotion_dict.get('fear',0) * 0.4 + emotion_dict.get('sadness',0) * 0.4 + emotion_dict.get('anger',0) * 0.2) * 100, 1)
}
top_trait = max(big5, key=big5.get)

insight = generate_insight(top_emotion, top_trait, stress_pct, stress_level)

print(f"📝 Text: '{text}'")
print(f"\n🎯 Top Emotion  : {top_emotion.upper()}")
print(f"⭐ Top Trait    : {top_trait.upper()}")
print(f"🔥 Stress Score : {stress_pct}% — {stress_level}")
print(f"\n💬 BehaviorLens Insight:")
print(f"   {insight}")
````

````python
def behaviorlens_full(text):
    print("\n" + "="*60)
    print("🧠 BEHAVIORLENS — FULL ANALYSIS")
    print("="*60)

    # Emotion
    results = emotion_model(text)
    if isinstance(results[0], list):
        results = results[0]
    results = sorted(results, key=lambda x: x['score'], reverse=True)
    emotion_dict = {r['label']: r['score'] for r in results}
    top_emotion = results[0]['label']
    top_emotion_score = results[0]['score'] * 100

    # Big 5
    big5 = {
        'Openness':          round((emotion_dict.get('joy',0) * 0.4 + emotion_dict.get('surprise',0) * 0.4 + emotion_dict.get('neutral',0) * 0.2) * 100, 1),
        'Conscientiousness': round((emotion_dict.get('neutral',0) * 0.5 + emotion_dict.get('joy',0) * 0.3 + (1 - emotion_dict.get('sadness',0)) * 0.2) * 100, 1),
        'Extraversion':      round((emotion_dict.get('joy',0) * 0.6 + emotion_dict.get('surprise',0) * 0.2 + emotion_dict.get('neutral',0) * 0.2) * 100, 1),
        'Agreeableness':     round((emotion_dict.get('joy',0) * 0.4 + emotion_dict.get('neutral',0) * 0.3 + (1 - emotion_dict.get('sadness',0)) * 0.3) * 100, 1),
        'Neuroticism':       round((emotion_dict.get('fear',0) * 0.4 + emotion_dict.get('sadness',0) * 0.4 + emotion_dict.get('anger',0) * 0.2) * 100, 1)
    }
    top_trait = max(big5, key=big5.get)

    # Stress
    stress_pct, stress_level = calculate_stress(emotion_dict)

    # Overall BehaviorLens Score
    positivity = emotion_dict.get('joy', 0) + emotion_dict.get('surprise', 0) * 0.5
    bl_score = round((positivity * 0.4 + (1 - stress_pct/100) * 0.4 + emotion_dict.get('neutral', 0) * 0.2) * 100, 1)

    # Insight
    insight = generate_insight(top_emotion, top_trait, stress_pct, stress_level)

    # Print results
    print(f"\n📝 Input: '{text}'")
    print(f"\n{'─'*60}")
    print(f"  🎯 Dominant Emotion  : {top_emotion.upper()} ({top_emotion_score:.1f}%)")
    print(f"  ⭐ Dominant Trait    : {top_trait.upper()} ({big5[top_trait]}%)")
    print(f"  🔥 Stress Level      : {stress_pct}% — {stress_level}")
    print(f"  💯 BehaviorLens Score: {bl_score}% — Overall Positivity")
    print(f"{'─'*60}")
    print(f"\n  💬 Insight:")
    print(f"     {insight}")
    print(f"\n  📊 All Emotions:")
    for r in results:
        bar = "█" * int(r['score'] * 25)
        print(f"     {r['label']:<12} {bar} {r['score']*100:.1f}%")
    print(f"\n  🧠 Big 5 Personality:")
    for trait, score in big5.items():
        bar = "█" * int(score / 5)
        print(f"     {trait:<20} {bar} {score}%")
    print("\n" + "="*60)

# TEST IT!
behaviorlens_full("I love exploring new ideas and meeting new people every day!")
behaviorlens_full("I am so scared and worried about my future.")
behaviorlens_full("I got rejected from the job I really wanted.")
````

````python
# Save all our functions to a python file
code = '''
from transformers import pipeline
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# Load model
emotion_model = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=True
)

def get_emotion(text):
    results = emotion_model(text)
    if isinstance(results[0], list):
        results = results[0]
    results = sorted(results, key=lambda x: x["score"], reverse=True)
    emotion_dict = {r["label"]: r["score"] for r in results}
    return results, emotion_dict

def get_big5(emotion_dict):
    return {
        "Openness":          round((emotion_dict.get("joy",0) * 0.4 + emotion_dict.get("surprise",0) * 0.4 + emotion_dict.get("neutral",0) * 0.2) * 100, 1),
        "Conscientiousness": round((emotion_dict.get("neutral",0) * 0.5 + emotion_dict.get("joy",0) * 0.3 + (1 - emotion_dict.get("sadness",0)) * 0.2) * 100, 1),
        "Extraversion":      round((emotion_dict.get("joy",0) * 0.6 + emotion_dict.get("surprise",0) * 0.2 + emotion_dict.get("neutral",0) * 0.2) * 100, 1),
        "Agreeableness":     round((emotion_dict.get("joy",0) * 0.4 + emotion_dict.get("neutral",0) * 0.3 + (1 - emotion_dict.get("sadness",0)) * 0.3) * 100, 1),
        "Neuroticism":       round((emotion_dict.get("fear",0) * 0.4 + emotion_dict.get("sadness",0) * 0.4 + emotion_dict.get("anger",0) * 0.2) * 100, 1)
    }

def get_stress(emotion_dict):
    stress = ((emotion_dict.get("fear",0) * 0.35) +
              (emotion_dict.get("sadness",0) * 0.25) +
              (emotion_dict.get("anger",0) * 0.25) +
              (emotion_dict.get("disgust",0) * 0.15) -
              (emotion_dict.get("joy",0) * 0.3) -
              (emotion_dict.get("neutral",0) * 0.1))
    stress = max(0, min(1, stress))
    stress_pct = round(stress * 100, 1)
    if stress_pct < 20:   level = "🟢 Very Low"
    elif stress_pct < 40: level = "🟡 Low"
    elif stress_pct < 60: level = "🟠 Moderate"
    elif stress_pct < 80: level = "🔴 High"
    else:                 level = "🚨 Very High"
    return stress_pct, level

def get_insight(top_emotion, top_trait, stress_pct):
    insights = {
        ("joy", "Extraversion"):          "This person is highly energetic and socially confident.",
        ("joy", "Openness"):              "This person is creative, curious and in a positive state of mind.",
        ("joy", "Agreeableness"):         "This person is warm, cooperative and easy to work with.",
        ("sadness", "Neuroticism"):       "This person may be going through emotional difficulty. Handle with care.",
        ("anger", "Neuroticism"):         "This person is frustrated and emotionally reactive right now.",
        ("fear", "Neuroticism"):          "This person is anxious and may need reassurance.",
        ("neutral", "Conscientiousness"): "This person is calm, focused and highly organized.",
        ("surprise", "Openness"):         "This person is open to new experiences and easily excited.",
    }
    key = (top_emotion.lower(), top_trait)
    insight = insights.get(key, f"This person shows strong {top_trait} with {top_emotion} as dominant emotion.")
    if stress_pct > 60:   insight += " ⚠️ High stress detected — may need support."
    elif stress_pct < 20: insight += " ✅ Overall mental state appears healthy and balanced."
    return insight
'''

with open('behaviorlens_engine.py', 'w') as f:
    f.write(code)

print("Engine saved! ✅")
````
